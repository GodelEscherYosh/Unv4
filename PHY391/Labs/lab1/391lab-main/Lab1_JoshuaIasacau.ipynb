We are now going to try out some of these python skills on a more realistic analysis example.

There are several ways to read a CSV file.  The brute force appraoach is to write some code to parse the input by hand, but usually a package like <tt>pandas</tt> is more convenient.  Examples can be found [here](https://pandas.pydata.org/pandas-docs/stable/index.html) but I have shown a simple example of reading a CSV file below.


```python
# Import pandas
import numpy as np
import pandas as pd

# Open and read the CSV file
galaxy_data = pd.read_csv('galaxydata.txt')

# Show the first few rows, useful to see all the column names
galaxy_data.head(3)
```

Pandas creates what is known as a DataFrame. We can access part of a DataFrame using the python dictionary and list syntax.  Here we show the first five elements of the galaxy type column.  More examples are in the [Pandas getting started tutorials](https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/index.html).

```python
galaxy_data['type'][:5]
```

We can also make selections on different columns of the DataFrame.  Here we show the absolute magnitude for all Sbc galaxies.

```python
# Note logical equality is two equals signs, a single equals sign is an assignment operator
sbc = galaxy_data[(galaxy_data['type'] == 'Sbc')]
print(sbc['mabs'])
```

Now lets read in some data and make some plots.

Make a sky map scatter plot of Declination vs. Right Ascension for all of the galaxies in the file.  Make sure to convert RA to degrees, and size your markers appropriately.  It is probably worth separating the data access from the DataFrame from the plotting, as one you read the data correctly, you won't need to do this again.

```python
# Read the data here, assigning the arrays of numbers 
# to their own variables to make things more clear
# Declenation vs Right Ascesion
# Right Ascension is on the x axis of a sky map scatter plot of Declination vs. Right Ascension.

# Right Ascension is a celestial coordinate that is analogous to longitude on Earth. It is measured in hours, minutes, and seconds, and ranges from 0h to 24h. Right Ascension is measured from the vernal equinox, which is the point where the Sun crosses the celestial equator from south to north.
dec = galaxy_data['de2000']
rasc = galaxy_data['al2000']
rasc = rasc*15
```

```python
# Make the plot here
plt.scatter(rasc, dec, s=2)
plt.xlabel("Right Ascension (hours)")
plt.ylabel("Declenation (degrees)")
plt.show()
```

Now make a color-color plot of (ut-bt) vs. (vt-kt)

```python
# Go ahead and add some more data access if necessary, then make the plot here
utbt = galaxy_data['ut'] - galaxy_data['bt']
vtkt = galaxy_data['vt'] - galaxy_data['kt']
plt.scatter(vtkt, utbt, s=5)
plt.xlabel('vt-kt (brightness)')
plt.ylabel("ut-bt (brightness)")
plt.show()
```

Now plot (bt-kt) versus the surface brightness (brief).  Do this separately for spiral and eliptical galaxies.

```python
elip= galaxy_data[(galaxy_data['type'] == 'E')]
spiral= galaxy_data[(galaxy_data['type'] != 'E')]

ebtkt = elip['bt'] - elip['kt']
sbtkt = spiral['bt'] - spiral['kt']

sb = spiral['brief']
esb = elip['brief']

#Now plot
plt.scatter(sb, sbtkt, color='blue', label = 'spiral', s=2)
plt.scatter(esb, ebtkt, color='red', label = 'eliptical', s=2)
plt.xlabel('surface brightness ')
plt.ylabel('bt-kt brightness')
plt.legend(title = 'Legend Key', loc = 'upper right', facecolor='gray')
plt.show

```

Enter some comments about what you observe:

The data points representing red, which are the eleptical stars, cover a lot less surface area over the same size. In this way, they are more dense. Likewise, spiral galaxies occupy a lot more area in the graph, meaning that they have more spread.

Finally make a histogram of the redshift (v) for spiral and elliptical galaxies.

```python
redshiftspiral = spiral['v']
redshiftElliptical = elip['v']

plt.hist(redshiftspiral, color='Blue', alpha=0.7, label='spiral')
plt.hist(redshiftElliptical, histtype='step', color='red', label='Eliptical')
plt.xlabel('Redshift')
plt.ylabel("Galaxy count")
plt.legend(title = 'Legend Key', loc = 'upper right', facecolor='gray')
plt.show
```
